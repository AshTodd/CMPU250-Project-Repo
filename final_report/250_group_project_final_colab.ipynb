# Cleaning Data

```python
#select only reports of Intimate partner violence

#Clean 'Intimate Relationship Flag' column
np.unique(df['Intimate Relationship Flag'])
df['Intimate Relationship Flag'] = df['Intimate Relationship Flag'].str.title()

#determine length of remaining data.
df_IPV = df.groupby(by='Intimate Relationship Flag').get_group('Yes')
len(df_IPV)
```

```python
#Check for data type consistency
df_IPV.info()
```

```python
# count null values in data set
df_IPV.isnull().sum()
```

```python
#clean IPV data, changing NAN values.

#changing NaNs in columns: Poverty, Median Income, and Unemployment to value 0
df_IPV['Poverty'] = df_IPV['Poverty'].fillna(0)
df_IPV['Median Income'] = df_IPV['Median Income'].fillna(0)
df_IPV['Unemployment'] = df_IPV['Unemployment'].fillna(0)

#remove certainly incorect age 1017.
df_IPV = df_IPV.drop(df_IPV[df_IPV['Suspect Reported Age'] == 1017].index)

#Replace NAN value in Age columns with adverage age.
df_IPV['Victim Reported Age'] = df_IPV['Victim Reported Age'].fillna( df_IPV['Victim Reported Age'].mean())
df_IPV['Suspect Reported Age'] = df_IPV['Suspect Reported Age'].fillna( df_IPV['Suspect Reported Age'].mean())

#change ages to integers
df_IPV['Victim Reported Age'] = df_IPV['Victim Reported Age'].astype(int)
df_IPV['Suspect Reported Age'] = df_IPV['Suspect Reported Age'].astype(int)

#Drop remain rows containg NAN using .dropna()
clean_df = df_IPV.dropna()

#change Sex to binary form, 0=female 1=male
clean_df['Suspect Sex'] = clean_df['Suspect Sex'].replace('MALE', 1).replace('FEMALE', 0)
clean_df = clean_df.drop(clean_df[clean_df['Suspect Sex'] == 'UNKNWN'].index)
clean_df['Victim Sex'] = clean_df['Victim Sex'].replace('MALE', 1).replace('FEMALE', 0)
clean_df = clean_df.drop(clean_df[clean_df['Victim Sex'] == 'UNKNWN'].index)

#change Report Date to datetime
clean_df['Report Date'] = pd.to_datetime(clean_df['Report Date'])
```

```python
#Check data type consitancy
clean_df.info()
```

```python
#Add Binary Column marking the report as a Felony (1) or DIR (0)

def offense(offense_type):
    if offense_type == "RAPE" or offense_type == 'FELONY ASSAULT':
        return 1
    else:
        return 0

clean_df['Felony Offence'] = clean_df['Offense Type'].apply(offense)
```

# Model

```python
#get last weeks reports
#mark which prcincts apeared in the last weeks high reports
#get this weeks precincts
#mark difference
```

```python
#helper, filters dataframe when given df year and month
def weekFilter(df, year, month, day):
    newdf = df.loc[df['Report Date'] > datetime.datetime(year=year, month=month, day=day)]
    newdf = newdf.loc[df['Report Date'] < datetime.datetime(year=year, month=month, day=day+7)]
    return newdf

#helper to ensure timing updates
def update_date(month, year):
  if month<12:
    return month+1, year
  return 1, year+1
```

```python
def get_week_data(month, year, df):

  #get curent weeks data
  cur_week = weekFilter(df, year, month, 1)
  cur_week['Report Count'] = cur_week['Incident Precinct Code'].map(cur_week['Incident Precinct Code'].value_counts())


  #update the dates for the next week
  next_month, next_year = update_date(month, year)

  #get data for the next week
  next_week = weekFilter(df, next_year, next_month, 8)

  #update cur_week with next weeks report data
  cur_week['Next Week Report Count'] = cur_week['Incident Precinct Code'].map(next_week['Incident Precinct Code'].value_counts())
  cur_week['Next Week Report Count'] = cur_week['Next Week Report Count'].fillna(0)

  #find top 10 precincts by report count for this week and next week
  top10_this_week = cur_week['Incident Precinct Code'].value_counts().head(10).index
  top10_next_week = next_week['Incident Precinct Code'].value_counts().head(10).index

  #mark whether each row is in top 10
  cur_week['Next Week Top 10 Prediction'] = cur_week['Incident Precinct Code'].isin(top10_this_week)
  cur_week['Next Week Top 10'] = cur_week['Incident Precinct Code'].isin(top10_next_week)

  return cur_week
```

```python
accuracy_by_month = {}
all_precinct_rows = []

year = 2020
for i in range(12):
  month = i+1
  cur_week = get_week_data(month, year, clean_df)
  cur_week_felony = get_week_data(month, year, felony_df)
  #group and aggregate to one line per precinct
  precinct_df = cur_week.groupby('Incident Precinct Code').agg({
      'Report Count': 'first',
      'Next Week Top 10 Prediction': 'first',
      'Next Week Report Count': 'first',
      'Next Week Top 10': 'first'
  }).reset_index()
 #group and aggregate to one line per precinct for felony data
  precinct_df_felony = cur_week_felony.groupby('Incident Precinct Code').agg({
      'Report Count': 'first',
      'Next Week Top 10 Prediction': 'first',
      'Next Week Report Count': 'first',
      'Next Week Top 10': 'first',

  }).reset_index()
  #group and aggregate to one line per precinct for felony data
  precinct_df_felony = precinct_df_felony.rename(columns={
    'Report Count':'Felony Report Count',
    'Next Week Top 10 Prediction': 'Next Week Felony Top 10 Prediction',
    'Next Week Report Count': 'Next Week Felony Report Count',
    'Next Week Top 10':'Next Week Felony Top 10',

    })
  precinct_df = pd.merge(precinct_df, precinct_df_felony)

  precinct_df['Month'] = month
  #precinct_df['Year'] = year

  all_precinct_rows.append(precinct_df)

  #split data
  train, test = sk.model_selection.train_test_split(precinct_df, test_size=0.2)

  X_train, X_test = train[['Report Count', 'Felony Report Count']], test[['Report Count', 'Felony Report Count']]
  y_train, y_test = train['Next Week Top 10'], test['Next Week Top 10']
  y_train_felony, y_test_felony = train['Next Week Felony Top 10'], test['Next Week Felony Top 10']

  # fit model
  model = sk.linear_model.LogisticRegression(max_iter = 1000)

  model = model.fit(X_train, y_train)
  model_felony = model.fit(X_train, y_train_felony)

  # report accuracy on test set
  y_pred = model.predict(X_test)

  #np.count_nonzero(y_pred == y_test)/len(y_test)
  acc = sk.metrics.accuracy_score(y_pred, y_test)
  acc_felony = sk.metrics.accuracy_score(y_pred, y_test_felony)
  accuracy_by_month[month] = acc, acc_felony
  #print(acc)

# Combine all into a single DataFrame
all_precinct_data = pd.concat(all_precinct_rows, ignore_index=True)
accuracy_by_month
```

```python
y_true = all_precinct_data['Next Week Top 10']
y_pred = all_precinct_data['Next Week Top 10 Prediction']

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Top 10', 'Top 10'])

# Plot
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix: DIR Hotspot Prediction')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()
```

```python
# 1. Compute monthly metrics
monthly_metrics = []
for month in sorted(all_precinct_data['Month'].unique()):
    df_m = all_precinct_data[all_precinct_data['Month'] == month]
    y_true = df_m['Next Week Top 10']
    y_pred = df_m['Next Week Top 10 Prediction']
    monthly_metrics.append({
        'Month': month,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, zero_division=0),
        'Recall': recall_score(y_true, y_pred, zero_division=0),
    })

metrics_df = pd.DataFrame(monthly_metrics)
print(metrics_df.to_string(index=False))

# 2. Plot each metric over the months
for metric in ['Accuracy', 'Precision', 'Recall']:
    plt.figure()
    plt.plot(metrics_df['Month'], metrics_df[metric], marker='o')
    plt.xlabel('Month')
    plt.ylabel(metric)
    plt.title(f'Monthly {metric} Trend')
    plt.xticks(metrics_df['Month'])
    plt.tight_layout()
    plt.show()
```
